__Ridge Regression and Lasso in python__

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5)          ## alpha = lamda
lm_r.fit(X_train_s,Y_train)
```

```python
Y_test_s = lm_r.predict(X_test_s)
```

```python
r2_score(Y_test,Y_test_s)       
##y_test is original variable and 2nd arrgument is predicted value 
```

```python
from sklearn.model_selection import validation_curve
```

```python
param_range = np.logspace(-2,8,100)     ## use this value for lemda and select best lemda value
param_range
```

```python
train_scores, test_scores = validation_curve(Ridge(), X_train_s, Y_train, param_name="alpha", param_range=param_range, scoring='r2')
```

```python
print(train_scores)
print(test_scores)
```

```python
train_mean = np.mean(train_scores, axis = 1)
test_mean = np.mean(test_scores, axis = 1)
```

```python
train_mean
```

```python
test_mean
```

```python
max(test_mean)
```

```python
sns.jointplot(x=np.log(param_range),y=test_mean)
```

```python
np.where(test_mean == max(test_mean))
```

```python
param_range[30]
```

```python
lm_r_best = Ridge(alpha = param_range[30])
```

```python
lm_r_best.fit(X_train_s,Y_train)
```

```python
r2_score(Y_test , lm_r_best.predict(X_test_s))
```

```python
r2_score(Y_train , lm_r_best.predict(X_train_s))
```
